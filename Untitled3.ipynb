```python
import torch
import torch.nn as nn
model = nn.Sequential(
nn.Linear(5, 6),
nn.ReLU(),
nn.Linear(6, 4),
nn.sigmoid())
x = torch.rand(10, 5)
z = model(x)
print(z)
```
