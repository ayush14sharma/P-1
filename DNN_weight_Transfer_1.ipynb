```python
import torch

input = torch.tensor([1.0, 8.0, 4.0])

#Hidden Layer 1
w1 = torch.tensor([[0.2, 0.3, 0.28],
                   [0.54, 0.25, -0.498]],requires_grad = True)

b1 = torch.tensor([-0.26, 0.07],requires_grad = True)

w2 = torch.tensor([[-0.1, 0.2],
                  [0.3, 0.2]],requires_grad = True)
b2 = torch.tensor([0.20163, 0.54], requires_grad = True)


weight = torch.tensor([
[0.2, 0.4,],
[0.3, 0.6,]],requires_grad = True)

def sigmoid(x):
    return 1 / ( 1 + torch.exp(-x) )

bias = torch.tensor([1.0, -0.8],requires_grad = True)

target = torch.tensor([0.0, 0.0])

learning_rate = 0.1

for epoch in range(1000):
    h1 = sigmoid(torch.matmul(w1, input) + b1)
    h2 = sigmoid(torch.matmul(w2, h1) + b2)
    output = torch.matmul(weight, h2) + bias
    final_output = sigmoid(output)
    loss = ((target - final_output)**2).mean()
    loss.backward( )
    with torch.no_grad():
        w1-= learning_rate*w1.grad
        b1 -= learning_rate*b1.grad
        w2 -= learning_rate*w2.grad
        b2 -= learning_rate*b2.grad
        weight -= learning_rate*weight.grad
        bias -= learning_rate*bias.grad

    w1.grad.zero_( )
    b1.grad.zero_( )
    w2.grad.zero_( )
    b2.grad.zero_( )
    weight.grad.zero_( )
    bias.grad.zero_( )
print("\n")
print("Final w1 =", w1)
print("Final b1 =", b1)
print("\n")
print("Final w2 =", w2)
print("Final b2 =", b2)
print("\n")
print("Final weight =", weight)
print("Final bias =", bias)
print("Final output =", final_output)
print("Final loss =", loss.item())
print("\n")
```
